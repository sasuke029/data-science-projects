### Now lets do Data Preprocessing 

### Data Preprocessing
Data preprocessing is a necessary step towards transforming raw data into a format which can be utilized by machine learning algorithms. It involves addressing different types of data (numerical as well as categorial), scaling numerical features, and encoding the multi-class target feature.

Feature Type Identification: We must identify numerically columns which are truly numerical and which columns are categorial, although given as integers (e.g., 'Marital status', 'Course' may be categories).

One-Hot Encoding: Categorical columns are converted into a numerical format using One-Hot Encoding. This creates new binary columns for each category, preventing the model from assuming any ordinal relation between categories.

Standard Scaling: Numerical columns are scaled by StandardScaler. This sets the data to mean 0 and standard deviation 1, which is important for feature magnitude-sensitive algorithms (e.g., SVM, KNN, Neural Networks).

ColumnTransformer and Pipeline: ColumnTransformer is utilized to perform different preprocessing tasks on several columns simultaneously. This is then integrated into a Pipeline for effective data transformation.

Target Encoding: Multi-class target variable ('Dropout', 'Graduate', 'Enrolled') is encoded into numerical form (e.g., 0, 1, 2) by using LabelEncoder and then converted again into one-hot encoded form (to_categorical) for the Neural Network.

Train-Test Split: The data are split into training and test sets to train the models and test them against unseen data. Stratified splitting is used to maintain the ratio of the target classes within the two sets, particularly for multi-class and possibly unbalanced data.


### Feature Selection with RFE (Recursive Feature Elimination)
Feature selection is the process of choosing a subset of relevant features for use in model construction. It helps in reducing overfitting, improving model accuracy, and reducing training time.

Recursive Feature Elimination (RFE) is a wrapper-type feature selection algorithm. It works by recursively training a model and eliminating the least important features, based on the model's coef_ or feature_importances_ attribute, until the desired number of features is reached.

In this project, we will use LogisticRegression (adapted for multi-class with multi_class='ovr') as the base estimator for RFE. We will select a subset of features that are most influential in predicting student academic outcomes.

```python
# Clean and prepare data
df_clean = df.dropna()
X = df_clean.drop(columns=['Target'])
y_class = df_clean['Target'].apply(lambda x: 1 if x == 'Graduate' else 0)
```

```python

# Scale features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python

# Train/test split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_class, test_size=0.3, random_state=42)
```

```python
# Baseline Logistic Regression
logreg = LogisticRegression(max_iter=1000)
logreg.fit(X_train, y_train)
y_pred_log = logreg.predict(X_test)
```

```python
# Polynomial Logistic Regression
degree = 2
poly = PolynomialFeatures(degree)
X_poly = poly.fit_transform(X_scaled)
X_train_p, X_test_p, _, _ = train_test_split(X_poly, y_class, test_size=0.3, random_state=42)
logreg_poly = LogisticRegression(max_iter=1000)
logreg_poly.fit(X_train_p, y_train)
y_pred_poly = logreg_poly.predict(X_test_p)
```

```python
# RFE with Logistic Regression
selector = RFE(LogisticRegression(max_iter=1000), n_features_to_select=10)
selector.fit(X_train, y_train)
X_train_rfe = selector.transform(X_train)
X_test_rfe = selector.transform(X_test)
logreg_rfe = LogisticRegression(max_iter=1000)
logreg_rfe.fit(X_train_rfe, y_train)
y_pred_rfe = logreg_rfe.predict(X_test_rfe)
```

```python
# KNN with GridSearch
knn_params = {'n_neighbors': [3, 5, 7, 9]}
knn = GridSearchCV(KNeighborsClassifier(), knn_params, scoring='accuracy', cv=5)
knn.fit(X_train, y_train)
y_pred_knn = knn.predict(X_test)
```

```python
# 5ï¸âƒ£ SVM
svm = SVC(probability=True)
svm.fit(X_train, y_train)
y_pred_svm = svm.predict(X_test)
```

```python
# Random Forest with GridSearch
rf_params = {'n_estimators': [50, 100], 'max_depth': [5, 10]}
rf = GridSearchCV(RandomForestClassifier(), rf_params, scoring='accuracy', cv=5)
rf.fit(X_train, y_train)
y_pred_rf = rf.predict(X_test)
```

```python
# 7ï¸âƒ£ Gradient Boosting
gb = GradientBoostingClassifier()
gb.fit(X_train, y_train)
y_pred_gb = gb.predict(X_test)
```

```python
# Neural Network tuning using GridSearchCV

param_grid = {
    'hidden_layer_sizes': [(50,), (100,), (100, 50), (100, 100, 50)],
    'activation': ['relu', 'tanh'],
    'solver': ['adam'],
    'alpha': [0.0001, 0.001, 0.01],
    'learning_rate': ['constant', 'adaptive']
}
mlp_grid = GridSearchCV(MLPClassifier(max_iter=1000, random_state=42), param_grid, scoring='accuracy', cv=5, n_jobs=-1)
mlp_grid.fit(X_train, y_train)
y_pred_mlp = mlp_grid.predict(X_test)

```

```python
# Evaluation metrics
results = pd.DataFrame({
    'Model': ['Logistic Regression', 'Polynomial Logistic', 'RFE Logistic', 'KNN', 'SVM', 'Random Forest', 'Gradient Boosting', 'Neural Network'],
    'Accuracy': [
        accuracy_score(y_test, y_pred_log),
        accuracy_score(y_test, y_pred_poly),
        accuracy_score(y_test, y_pred_rfe),
        accuracy_score(y_test, y_pred_knn),
        accuracy_score(y_test, y_pred_svm),
        accuracy_score(y_test, y_pred_rf),
        accuracy_score(y_test, y_pred_gb),
        accuracy_score(y_test, y_pred_mlp)
    ],
    'F1 Score': [
        f1_score(y_test, y_pred_log),
        f1_score(y_test, y_pred_poly),
        f1_score(y_test, y_pred_rfe),
        f1_score(y_test, y_pred_knn),
        f1_score(y_test, y_pred_svm),
        f1_score(y_test, y_pred_rf),
        f1_score(y_test, y_pred_gb),
        f1_score(y_test, y_pred_mlp)
    ],
    'AUC': [
        roc_auc_score(y_test, logreg.predict_proba(X_test)[:, 1]),
        roc_auc_score(y_test, logreg_poly.predict_proba(X_test_p)[:, 1]),
        roc_auc_score(y_test, logreg_rfe.predict_proba(X_test_rfe)[:, 1]),
        roc_auc_score(y_test, knn.predict_proba(X_test)[:, 1]),
        roc_auc_score(y_test, svm.predict_proba(X_test)[:, 1]),
        roc_auc_score(y_test, rf.predict_proba(X_test)[:, 1]),
        roc_auc_score(y_test, gb.predict_proba(X_test)[:, 1]),
        roc_auc_score(y_test, mlp_grid.predict_proba(X_test)[:, 1])
    ]
})

print(results)
```

```python
# Plot Accuracy and F1 Score
# ðŸ“Š Plot Accuracy and F1 Score
plt.figure(figsize=(12, 6))
plt.bar(results['Model'], results['Accuracy'], alpha=0.7, label='Accuracy')
plt.bar(results['Model'], results['F1 Score'], alpha=0.7, label='F1 Score')
plt.title('Model Accuracy and F1 Score Comparison')
plt.ylabel('Score')
plt.legend()
plt.xticks(rotation=20)
plt.tight_layout()
plt.show()
```

```python
# Plot AUC
plt.figure(figsize=(12, 6))
plt.bar(results['Model'], results['AUC'], alpha=0.7, color='green', label='AUC')
plt.title('Model AUC Comparison')
plt.ylabel('AUC')
plt.xticks(rotation=20)
plt.tight_layout()
plt.show()
```

```python

# Plot Confusion Matrix and ROC Curve for Tuned Neural Network
from sklearn.metrics import ConfusionMatrixDisplay, RocCurveDisplay

ConfusionMatrixDisplay.from_estimator(mlp_grid, X_test, y_test)
plt.title('Neural Network Confusion Matrix')
plt.show()

RocCurveDisplay.from_estimator(mlp_grid, X_test, y_test)
plt.title('Neural Network ROC Curve')
plt.show()

print(f"Best Neural Network Parameters: {mlp_grid.best_params_}")

```
